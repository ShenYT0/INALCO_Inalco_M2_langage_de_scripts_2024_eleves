```python
from langdetect import detect
from nltk.tokenize import sent_tokenize

def detect_english(text:str) -> bool:
    try:
        return detect(text) == "en"
    except:
        return True

def remove_english_lines(text:str) -> str:
    lines = sent_tokenize(text)
    for line in lines:
        if detect_english(line):
            text = text.replace(line, "")
    return text

```

* Ã©crivez une fonction prenant en entrÃ©e le texte d'un roman et effectuant un split sur les sÃ©quences du type deux sauts de ligne, une ligne avec du texte, deux sauts de ligne, et qui renvoie un dictionnaire associant la chaÃ®ne de caractÃ¨re contenue entre les sauts de ligne et lui associe le texte avant la prochaine sÃ©paration


```python
from typing import Dict
import re

def is_valid_roman_numeral(s):
    pattern = r'^(M{0,4})(CM|CD|D?C{0,3})(XC|XL|L?X{0,3})(IX|IV|V?I{0,3})$'
    return bool(re.match(pattern, s))

# SalammbÃ´ why you so unique?
def has_six_spaces(text:str) -> bool:
    leading_spaces = len(text) - len(text.lstrip())
    return leading_spaces == 6

def split_in_chapters(text:str) -> Dict[str,str]:
    chapters = {}
    lines = text.splitlines()
    currentId = None
    for i,line in enumerate(lines):
        if has_six_spaces(line):
            line = line.strip()
        isId = False
        if i>2 and i<len(lines)-2:
            # if lines[i-1] == "" and lines[i+1] == "" and lines[i-2] == "" and lines[i+2] == "":
            if lines[i-1] == "" and lines[i+1] == "" and is_valid_roman_numeral(line) and line != "":
                isId = True
        if isId:
            currentId = line
            while currentId in chapters:
                currentId = "." + currentId
            chapters[currentId] = ""
            # if line not in chapters:
            #     currentId = line
            #     chapters[currentId] = ""
            # else:
            #     currentId = '.' + line
            #     chapters[currentId] = ""
        elif line != "" and currentId is not None:
            chapters[currentId] += line
                

    return chapters

chaptersES = split_in_chapters(puredES)
```

```python
cleanDict = clean_text(dictionnaire)

def divise_dict(text:str) -> Dict[str,str]:
    lines = text.splitlines()
    result = {}
    currentWord = None

    for line in lines:
        if ":" in line:
            parts = line.split(":")
            firstWord = parts[0].split(" ")[0]
            if firstWord.isupper():
                currentWord = parts[0]
                result[currentWord] = "".join(parts[1:])
            elif currentWord is not None:
                result[currentWord] += line
        elif currentWord is not None:
            result[currentWord] += line

    return result
    
dictDict = divise_dict(cleanDict)
```
